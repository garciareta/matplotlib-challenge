```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
path1 = os.path.join("data", "mouse_drug_data.csv")
path2 = os.path.join("data", "clinicaltrial_data.csv")

# Read the Mouse and Drug Data and the Clinical Trial Data
df1 = pd.read_csv(path1)
df2 = pd.read_csv(path2)

# Combine the data into a single dataset
df = df2.merge(df1, how ="left", on='Mouse ID')

# Display the data table for preview
```

```python
df
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
# Convert to DataFrame
# Preview the DataFrame
tumor = df.groupby(["Drug", "Timepoint"], as_index=False).agg({"Tumor Volume (mm3)":["mean"]})
tumor.columns = ["Drug", "Timepoint","Tumor Volume (mm3)"]
tumor


```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
# Convert to DataFrame
# Preview DataFrame

tumerr = df.groupby(["Drug", "Timepoint"], as_index = False).agg({"Tumor Volume (mm3)":["sem"]})
tumerr.columns = ["Drug", "Timepoint", "Tumor Volume (mm3) sem"]
tumerr



```

```python
# Minor Data Munging to Re-Format the Data Frames
# Preview that Reformatting worked
tumor_piv = pd.pivot_table(df, values='Tumor Volume (mm3)',columns=['Drug'],index=['Timepoint'], aggfunc='mean')
tumor_piv

```

```python
tumor_err_piv = pd.pivot_table(df, values='Tumor Volume (mm3)',columns=['Drug'],index=['Timepoint'], aggfunc='sem')
tumor_err_piv
```

```python
# Generate the Plot (with Error Bars)
tumor_piv.plot(kind='line', figsize = (12, 6), grid=True, yerr = tumor_err_piv, marker='o', linestyle="--")
plt.title("Tumor response to Treatment")
plt.xlabel('Time (Days)')
plt.ylabel('Tumor Volume (mm3)')
plt.legend(loc='upper left')

# Save the Figure
plt.savefig("Tumor response to Treatment BG.png")
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
# Convert to DataFrame
# Preview DataFrame
metamean = df.groupby(["Drug", "Timepoint"]).agg({"Metastatic Sites":["mean"]})
metamean.columns = ["Metastatic Sites"]
metamean
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
# Convert to DataFrame
# Preview DataFrame
meterr = df.groupby(["Drug", "Timepoint"]).agg({"Metastatic Sites":["sem"]})
meterr.columns = ["Metastatic Sites"]
meterr
```

```python
# Minor Data Munging to Re-Format the Data Frames
# Preview that Reformatting worked
meta_piv = pd.pivot_table(df, values="Metastatic Sites",columns=["Drug"],index=["Timepoint"], aggfunc="mean")
meta_piv
```

```python
meterr_piv = pd.pivot_table(df, values="Metastatic Sites",columns=["Drug"],index=["Timepoint"], aggfunc="sem")
meterr_piv
```

```python
# Generate the Plot (with Error Bars)
meta_piv.plot(kind="line", figsize = (12, 6), title ="Metastatic Spread During Treatment", grid=True, yerr = meterr_piv, marker='o', linestyle="--")
plt.title("Metastatic Spread During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Metastatic Sites")

# Save the Figure
plt.savefig("Metastatic Spread During Treatment BG.png")

# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
# Convert to DataFrame
# Preview DataFrame
mcount = df.groupby(["Drug", "Timepoint"]).agg({"Mouse ID":["count"]})
mcount.columns = ["Mouse Count"]
mcount
```

```python
# Minor Data Munging to Re-Format the Data Frames
# Preview the Data Frame
count_piv = pd.pivot_table(df, values='Mouse ID',columns=['Drug'],index=['Timepoint'], aggfunc=('count'))
survival_rate = count_piv
count_piv
```

```python
micecount = count_piv.loc[0,:]
for column in count_piv:
    survival_rate[column] = count_piv[column] / micecount[column] * 100
    
survival_rate.round()
```

```python
# Generate the Plot (Accounting for percentages)
survival_rate.plot(kind="line", figsize = (12, 6), title ="Survival Rate During Treatment", grid=True, marker='X', linestyle="--")
plt.title("Survival Rate During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate")

# Save the Figure
plt.savefig("Survival Rate During Treatment BG.png")

# Show the Figure
plt.show()
```
